# Basic 1-dimensional ODE solution

```python
# This is a standard initalisation code - no need to change this
# It loads basic fast array code, plot library and ODE solver
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
%matplotlib inline
%config InlineBackend.figure_formats={'retina','png'}
```

## System definition

To solve the equation
$$
\frac{dx}{dt} = f(x)
$$
define the function $f$ here.

**Note**, that the function can explicitly depend on the free variable, to solve the *non-autonomous* equation $\dot{x}=f(t,x)$, but according to our convention this is no longer 1dim ODE, but 2dim ODE.

```python
def f(t, x):
    return math.sin(x)
```

## Basic analysis

First, let us plot the function to check and analyse what to expect.

```python
xrange = np.linspace(-6.3, 6.3, num=100) # Set the range of x
t = 0
fx = np.array([f(t, x) for x in xrange]) # Calculate the function values

fig1, ax1 = plt.subplots(1) # get figure and "axes". They can be reused later.

ax1.plot(xrange, fx) # Plot!
ax1.set_xlabel("$x$")
ax1.set_ylabel("$f(x)$")
ax1.grid()
```

## Solution

Solve the equations over given time range with specified intial condition

```python
tspan = (0.0, 10.0) # Define the required time range here
t_eval = np.linspace(tspan[0], tspan[1], num=100) # Make reported solution "smooth"
x0 = [1.0]          # Set the initial condition x(0)

xs = solve_ivp(f, tspan, x0, t_eval=t_eval)

fig, ax = plt.subplots(1)

ax.plot(xs.t, xs.y[0]) # Plot!
ax.set_xlabel("$t$")
ax.set_ylabel("$x$")
ax.grid()
```

Let us try for another one (we add to the previous plot by reusing `ax`)

```python
tspan = (0.0, 10.0) # Define the required time range here
t_eval = np.linspace(tspan[0], tspan[1], num=100) # Make reported solution "smooth"
x0 = [4.0]          # Set the initial condition x(0)

xs = solve_ivp(f, tspan, x0, t_eval=t_eval)

ax.plot(xs.t, xs.y[0])
fig # This actually shows the picture
```
