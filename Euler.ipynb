# Euler method

```
# This is a standard initalisation code - no need to change this
# It loads basic fast array code, plot library and ODE solver
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
%matplotlib inline
%config InlineBackend.figure_formats={'retina','png'}
```

```
def f(t, x):
    return x
```

```
## initial time and position
t = 0
x = 1

## step
h = 0.1

## Lists to store the solution, and store thed initial point
ti = [t]
xi = [x]

while t < 1.0:
    x = x+h*f(t, x)
    t = t+h
    xi.append(x)
    ti.append(t)

print("ti=", ti)
print("xi=", xi)
```

```
fig, ax = plt.subplots(1)

ax.scatter(ti, xi, label="Euler")
ax.set_xlabel("$t$")
fig.legend()
```

```
xexact = np.exp(ti)

ax.plot(ti, xexact, label="Exact solution")
fig.legend()
fig
```
